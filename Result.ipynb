```python
import numpy as np

x = np.arange(0,3.2,0.2)
y1 = np.load("./exact.npy")
y2 = np.load("./ori_once.npy")
y3 = np.load("./ori_mean.npy")
y4 = np.load("./mit.npy")
```

```python
from matplotlib import pyplot as plt

plt.plot(x, y1, label=r"Exact")
plt.plot(x, y2,label=r"Original")
plt.plot(x, y3,label=r"Mean Original")
plt.plot(x, y4, label=r"Mitigated")

plt.xlabel(r'Time $t$', fontsize="12")
plt.ylabel(r'Local magnetization', fontsize="12")
plt.legend()

plt.show()
```
